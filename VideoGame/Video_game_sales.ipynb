# Corelation between genre and sales

```python
data_gen
```

```python
sns.heatmap(data_gen,cmap="coolwarm", annot = True)
```

# How many games had been released in a particular genre in  each year

```python
temp=pd.crosstab(data['Year'],data['Genre']) #crosstab is used to count the relation between the rows and colunmns 
```

```python
temp
```

Above dataframe gives the count of individual genre released in each year.

```python
fig,ax=plt.subplots(2,1,figsize=(10,8))
top=data['Year'].value_counts().nlargest(10)
least=data['Year'].value_counts().nsmallest(10)

ax1 = sns.barplot(x=top.index,y=top.values,ax=ax[0])
ax2 = sns.barplot(x=least.index,y=least.values,ax=ax[1])
ax1.set(xlabel='year', ylabel='No of Releases')
ax2.set(xlabel='year', ylabel='No of Releases')
```

2009 and 2008 have highest number of releases, 2020 and 2017 has lowest number of releases.

# Represenatation of Sales

```python
sales=pd.Series({'NA':data.NA_Sales.mean(),
'EU':data.EU_Sales.mean(),
'JP':data.JP_Sales.mean(),
'OTHER':data.Other_Sales.mean(),
'GLOBAL':data.Global_Sales.mean()})
```

```python
sales
```

```python
plt.figure(figsize=(10,8))
plt.pie(sales,labels=sales.index, autopct = '%1.1f%%', wedgeprops={'width': 0.3})
plt.legend()
```
